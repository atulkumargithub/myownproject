```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.title("Sales")
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np. array([3, 8, 1, 10])

plt.bar(x, y, color='Red', width=0.1)
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np. array([3, 8, 1, 10])

plt.barh(x, y, height=0.1)
plt.show()
```

```python
x = np.random.normal(70, 8, 150)

plt.hist(x)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
myexplode = [0.2, 0, 0,0] 

#plt.pie(y, labels=mylabels, explode=myexplode, shadow=True
plt.pie(y, labels=mylabels, shadow=True)
plt.legend(title= "Total Sales")
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
myexplode = [0.2, 0, 0,0] 

plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
#plt.pie(y, labels=mylabels, shadow=True)
plt.legend(title= "Total Sales")
plt.show()
```
